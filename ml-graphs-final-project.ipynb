```python
import torch
import os

print(f"PyTorch Version: {torch.__version__}")
print(f"CUDA Version: {torch.version.cuda}")

# 1. Uninstall the broken packages
os.system("pip uninstall -y torch-scatter torch-sparse torch-cluster torch-spline-conv torch-geometric")

# 2. Construct the URL for the correct wheels
# We strip the "+cu1xx" part for the variable, but keep it for the URL if needed.
torch_ver = torch.__version__.split('+')[0]  # e.g., '2.1.0'
cuda_ver = torch.version.cuda.replace('.', '') # e.g., '121'

# URL format: https://data.pyg.org/whl/torch-{torch_ver}+cu{cuda_ver}.html
# Note: If on CPU only, usage is different, but for Colab GPU:
wheel_url = f"https://data.pyg.org/whl/torch-{torch.__version__}.html"

print(f"Installing from: {wheel_url}")

# 3. Reinstall with the correct link
# Note: We use os.system to run shell commands from python
os.system(f"pip install torch-scatter torch-sparse torch-cluster torch-spline-conv -f {wheel_url}")
os.system("pip install torch-geometric")

print("Done. Please Restart Runtime (Runtime > Restart Session) and run your imports again.")
```
